```python
import cv2
import sys

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('Camera open failed')
    sys.exit()
    
file_path = '../fig/my_images/'

idx = 0
while True:
    ret, frame = cap.read()
    file_name = 'face_{0:04d}.jpg'.format(idx)
    if not ret:
        print('frame with failed')
        break
    
    cv2.imshow('frame', frame)
#     print(frame.dtype, frame.shape)
    saveFile = file_path + file_name
    print(saveFile)
    cv2.imwrite(saveFile, frame)

    if cv2.waitKey(10) == 27:
        break
    idx += 1
        
cap.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
import sys
import cv2
import pandas as pd
import glob
```

```python
img_paths = glob.glob('../fig/my_images/*.*')

save_path = '../fig/my_images/not/'
## caffe
model = '../ssd_faceDetector/res10_300x300_ssd_iter_140000_fp16.caffemodel'
config = '../ssd_faceDetector/deploy.prototxt.txt'

# ## tensorflow
# model = './ssd_faceDetector/opencv_face_detector_uint8.pb'
# config = './ssd_faceDetector/opencv_face_detector.pbtxt.txt'

face_net = cv2.dnn.readNet(model, config)
if face_net.empty():
    print('Model read failed')
    sys.exit()
    
idx = 0
for img_path in img_paths:
    img = cv2.imread(img_path)
    print(img_path)
#     print(img.shape)
#     break
    blob = cv2.dnn.blobFromImage(img, 1, (300, 300), (104, 117, 123), swapRB=False)
    face_net.setInput(blob)
    outs = face_net.forward()
    
    detect = outs[0, 0, :, :]
    h, w = img.shape[:2]
    
    save_img = []
    for i in range(detect.shape[0]):
        confidence = detect[i, 2]
        if confidence> 0.5:
            x1 = int(detect[i, 3] * w)
            y1 = int(detect[i, 4] * h)
            x2 = int(detect[i, 5] * w)
            y2 = int(detect[i, 6] * h)
            
            file_name = 'face_{0:04d}.jpg'.format(idx)
            idx += 1
            save_img = img[y1:y2, x1:x2, :]
            cv2.imwrite(save_path + file_name, save_img)
```
